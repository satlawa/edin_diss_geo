```python
from osgeo import gdal, ogr

# Define pixel_size and NoData value of new raster
pixel_size = 1
NoData_value = 0

# Filename of input OGR file
vector_fn = '/home/philipp/Data/edin_diss/dringlich.shp'

# Filename of the raster Tiff that will be created
raster_fn = '/home/philipp/Data/edin_diss/test.tif'

# Open the data source and read in the extent
source_ds = ogr.Open(vector_fn)
source_layer = source_ds.GetLayer(0)
SR = source_layer.GetSpatialRef().ExportToWkt()
print(source_layer.GetName())
x_min, x_max, y_min, y_max = source_layer.GetExtent()
print(x_min, x_max, y_min, y_max)

# Create the destination data source
x_res = int((x_max - x_min) / pixel_size)
y_res = int((y_max - y_min) / pixel_size)
target_ds = gdal.GetDriverByName('GTiff').Create(raster_fn, x_res, y_res, 1, gdal.GDT_Byte)
target_ds.SetProjection(proj)
target_ds.SetGeoTransform((x_min, pixel_size, 0, y_max, 0, -pixel_size))
band = target_ds.GetRasterBand(1)
band.SetNoDataValue(NoData_value)

# Rasterize
layer_list = ['Nutzdringl']

gdal.Rasterize(raster_fn, vector_fn, format='GTiff', outputBounds=[x_min, y_min, x_max, y_max], outputSRS=SR, xRes=x_res, yRes=y_res, burnValues=[2], layers=layer_list)

target_ds = None
vector_fn = None
source_layer = None
source_ds = None
```
