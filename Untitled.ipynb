```python
#####

#Bounding Box Filter

#New in version 0.1.0.

#The bounding box filter only loads data that intersects with the bounding box.

bbox = (
    1031051.7879884212, 224272.49231459625, 1047224.3104931959, 244317.30894023244
)
gdf = geopandas.read_file(
    geopandas.datasets.get_path("nybb"),
    bbox=bbox,
)


#####
```

```python
# get extend
x_min, y_min, x_max, y_max = vector_data.geometry.total_bounds

pixel_size = 1

# Create the destination data source
x_res = int((x_max - x_min) / pixel_size)
y_res = int((y_max - y_min) / pixel_size)

print(x_res, y_res)
```

```python
out_grid = make_geocube(
    vector_data = vector_data,
    measurements = ['Nutzdringl'],
    resolution=(x_res, y_res),
)
#    output_crs="epsg:31287",
```

```python
out_grid.Nutzdringl.rio.to_raster("lsad_num.tif")
```
